```python
def cw(n, w):
    lis=[]
    s="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    for i in range(len(w)): #0, 1, 2
        lis.append(s.index(w[i]))
        
    res=""
    for i in lis:
        #if
        try:
            res+=s[i+n]
        #else
        except:
            res+=s[i+n-len(s)] # s[23+5-26] => S[2] 
    
    return res

print(cw(5,"CAT"))
```
